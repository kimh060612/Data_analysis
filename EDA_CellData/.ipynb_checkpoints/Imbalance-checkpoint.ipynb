```python
length = len(lines)

histogram = np.zeros(19)

for i in range(1, length):
    labels = lines['Label'][i].split('|')
    for con in labels:
        histogram[int(con)] += 1

print(histogram)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,5), dpi= 100)
ax = fig.subplots()

x = np.linspace(0, 18, 19)
x = [ str(int(num)) for num in x ]
ax.bar(x, histogram)
```

```python
path = '/data/CellData/train/'
images = [path + imgName['microtubules'], path + imgName['nucleus'], path + imgName['endoplasmic_reticulum'], path + imgName['protein_of_interest']]
```

```python
from PIL import Image

def image_to_arrays(path):
    
    image_arrays = list()
    for image in path:
        array = np.asarray(Image.open(image))
        image_arrays.append(array)
        
    return image_arrays

def get_blended_image(images): 
    
    blended_array = np.stack(images[:-1], 2)
    
    blended_image = Image.fromarray( np.uint8(blended_array) )
    return blended_image

image_set = image_to_arrays(images)
image_blended = get_blended_image(image_set)
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100)
imshow(image_blended)
```
